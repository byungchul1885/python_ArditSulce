### 데이타프레임 생성

```python
import pandas as pd
df = pd.read_csv("supermarkets.csv")
df.set_index("Address", inplace = True)
df
```

### 열 추가

```python
df["Continent"] = "North America"
df
```

```python
df["Continent"] = df["Country"] + "," + " North America"
df
```

### 행 추가
#### 열과 행 바꾸기

```python
df_t = df.T
df_t
```

#### 한 행 삽입

```python
df_t["My address"] = ["4", "Seoul", "Korea", "Repulbic of Korea", "bccho", "23", "hanbando"]
df_t
```

#### 열과 행 다시 바꾸기

```python
df = df_t.T
df
```
